```python
#single-valued landscape
from scipy import integrate
from scipy.integrate import quad


pop_size=10**5
N=pop_size
top_4_u=10**-5
u=top_4_u
s=0.05
alpha=2 #for the double mutant
f_max=0.4999999


#average number of double mutant clone detected in a person
def prob_double_mutant_function(t, VAF_lim, f_max): 
    f_min=VAF_lim
    n_star=(np.exp(alpha*s*t)-1)/(alpha*s)
    f_star=n_star/(2*(N+n_star))
    if VAF_lim<f_star:
        if f_max>f_star:
            f_max=f_star
    f_spectrum= lambda f: (2*N/(1-2*f))*(1+(2*f/(1-2*f)))*(alpha*N*u**2/(alpha*s*(2*N*f/(1-2*f))+1))*((np.exp(s*t)/(alpha*s*(2*N*f/(1-2*f))+1)**(1/alpha))-1)
    result = integrate.quad(f_spectrum, f_min,f_max)[0]
    
    if VAF_lim>f_star:
        result=0

    return result

#average number of separate mutations detected in a person
def prob_multiple_singles_function(t, VAF_lim, f_max): 
    n_tilde=(np.exp(s*t)-1)/s
    f_min=VAF_lim
    n_star=(np.exp(alpha*s*t)-1)/(alpha*s)
    f_star=n_star/(2*(N+n_star))
    if VAF_lim<f_star:
        if f_max>f_star:
            f_max=f_star
    f_spectrum= lambda f: (2*N/(1-2*f))*(1+(2*f/(1-2*f)))*N*u/(2*N*f/(1-2*f))*np.exp(-(2*N*f/(1-2*f))/n_tilde)
    result = integrate.quad(f_spectrum, f_min,f_max)[0]
    
    if VAF_lim>f_star:
        result=0

    return result

plt.close('all')
f, ax1 = plt.subplots(1, 1, figsize=(7*3,10))
plt.subplots_adjust(hspace = 0.7,wspace=0.35)  
font = {'weight' : 'normal','size'   : 16}
matplotlib.rc('font', **font)


linear_age=np.linspace(0,200,100)
ymin=10**-12
ymax=10**1

VAF_lim=0.001
# at least one double mutant
prob_double_mutant=[(1-np.exp(-prob_double_mutant_function(t, VAF_lim, f_max)))  for t in linear_age]  
#at least two separate mutations
prob_separate_mutations=[(1-np.exp(-prob_multiple_singles_function(t, VAF_lim, f_max))-\
                         prob_multiple_singles_function(t, VAF_lim, f_max)*np.exp(-prob_multiple_singles_function(t, VAF_lim, f_max))) for t in linear_age]   #includes 2 or more separate mutations




ax1.plot(linear_age, prob_double_mutant, color=green5, label='Detectable multiple mutants with VAF limit at 0.001')
ax1.plot(linear_age, prob_separate_mutations, color=blue1, label='Dectectable multiple separate mutations with VAF limit at 0.001')

VAF_lim=0.05



# at least one double mutant
prob_double_mutant=[(1-np.exp(-prob_double_mutant_function(t, VAF_lim, f_max)))  for t in linear_age]  
#at least two separate mutations
prob_separate_mutations=[(1-np.exp(-prob_multiple_singles_function(t, VAF_lim, f_max))-\
                         prob_multiple_singles_function(t, VAF_lim, f_max)*np.exp(-prob_multiple_singles_function(t, VAF_lim, f_max))) for t in linear_age]   #includes 2 or more separate mutations




ax1.plot(linear_age, prob_double_mutant, color=green3, label='Detectable multiple mutants with VAF limit at 0.05')
ax1.plot(linear_age, prob_separate_mutations, color=blue3, label='Dectectable multiple separate mutations with VAF limit at 0.05')


VAF_lim=0.1


# at least one double mutant
prob_double_mutant=[(1-np.exp(-prob_double_mutant_function(t, VAF_lim, f_max)))  for t in linear_age]  
#at least two separate mutations
prob_separate_mutations=[(1-np.exp(-prob_multiple_singles_function(t, VAF_lim, f_max))-\
                         prob_multiple_singles_function(t, VAF_lim, f_max)*np.exp(-prob_multiple_singles_function(t, VAF_lim, f_max))) for t in linear_age]   #includes 2 or more separate mutations






ax1.plot(linear_age, prob_double_mutant, color=green2, label='Detectable multiple mutants with VAF limit at 0.1')
ax1.plot(linear_age, prob_separate_mutations, color=blue5, label='Dectectable multiple separate mutations with VAF limit at 0.1')

handles, labels = ax1.get_legend_handles_labels()
# sort both labels and handles by labels
labels, handles = zip(*sorted(zip(labels, handles), key=lambda t: t[0]))
#ax.legend(handles, labels)
ax1.legend(handles, labels)
ax1.set_ylabel('Predicted Rate')
ax1.set_yscale('log')
ax1.set_xlabel('Age')  #assuming 1 generation per year
ax1.set_title('Single valued fitness landscape (s=5%)')
ax1.set_ylim(ymin,ymax) 






    
#plt.savefig('Predicted_detectable_rate_sinlge_valued_fitness_landscape_5_percent.pdf', bbox_inches='tight')
```
