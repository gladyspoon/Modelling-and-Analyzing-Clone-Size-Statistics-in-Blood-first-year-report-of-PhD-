```python
# Functions for calculating prevalence of double mutants (i.e. 2 mutations in 1 clone)
def mu_of_s(b, d, s, u): #mutation rate across an exponential distribution of s
    return (u/normalization)*np.exp(-((abs(s))/d)**b)

def expected_density_double_mutants_s_dist(N, s1, s2, t, f_min, b, d, u): #s1 = fitness of 1st mutation, s2 = fitness of 2nd mutation
    combined_s=s1+s2 #sum of the fitness effects of the 2 mutations
    alpha=combined_s/s1 #ratio of combined fitness to the original fitness (fitness of s1)
    mu1 = mu_of_s(b, d, s1, u) #mutation rate for 1st mutation with exponential distribution of s
    mu2 = mu_of_s(b, d, s2, u) #mutation rate for 2nd mutation with exponential distribution of s
    sigh = (np.exp(combined_s*t)-1)/(2*N*combined_s)
    if sigh <0.5:
        f_max = sigh
        expected_density = ((N*mu1*mu2)/s1)*(((alpha*np.exp(s1*t))*(1/(((2*N*combined_s*f_min)+1)**(1/alpha))-\
                                                                    1/(((2*N*combined_s*sigh)+1)**(1/alpha)))+\
                                                np.log(((2*N*combined_s*f_min)+1)/((2*N*combined_s*sigh)+1)) ))
    else:
        f_max = 0.4999
        expected_density = ((N*mu1*mu2)/s1)*(((alpha*np.exp(s1*t))*(1/(((2*N*combined_s*f_min)+1)**(1/alpha))-\
                                                                    1/(((2*N*combined_s*0.5)+1)**(1/alpha)))+\
                                                np.log(((2*N*combined_s*f_min)+1)/((2*N*combined_s*0.5)+1))))
        
    return (expected_density)
    
def after_integration_wrt_s1(N, s2, t, f_min, b, d, u): #integrate over exponential distribution of s (for 1st mutation)
    return integrate.quad(lambda s1: expected_density_double_mutants_s_dist(N, s1, s2, t, f_min, b, d, u), 0.02, s_max)[0]
    
def after_integration_wrt_s1_s2(N, t, f_min, b, d, u): #integrate over exponential distribution of s (for 2nd mutation)
    return integrate.quad(lambda s2: after_integration_wrt_s1(N, s2, t, f_min, b, d, u), 0.02, s_max)[0]

def prevalence_double_mutants(N, t, f_min, b, d, u):
    expected_number_double_mutant_clones=after_integration_wrt_s1_s2(N, t, f_min, b, d, u)
    return 1.0-np.exp(-expected_number_double_mutant_clones)
```
