```python
#Young 2016: one parameter optimization
datafilename='all_studies_not_trimmed_all_genes_multiple_mutations.csv'

# Look at the 10 most commonly mutated genes
#  DNMT3A: all coding exons
#  TET2: all coding exons
#  ASXL1: exon 12
#  JAK2: exon 12 and 14
#  TP53: all coding exons
#  SF3B1: exons 13 - 16
#  SRSF2: exon 1
#  IDH2: exon 4
#  KRAS: exons 2 - 3
#  CBL: exons 8 - 9

target_study='Young 2016'
num_data_pts=5
x_vector=range(num_data_pts)
total_pop_num=20
counts_6_mut=0
counts_5_mut=0
counts_4_mut=0
counts_3_mut=0
counts_2_mut=0
counts_1_mut=0
base_number_6_mut_lessthan70=0
base_number_5_mut_lessthan70=0
base_number_4_mut_lessthan70=0
base_number_3_mut=0
base_number_2_mut=0
base_number_1_mut=0

plt.close('all')
f, ax1 = plt.subplots(1, 1, figsize=(7*3,10))
plt.subplots_adjust(hspace = 0.7)  
font = {'weight' : 'normal','size'   : 16}
matplotlib.rc('font', **font)
VAF_lim=0.0011

linear_age=np.array(np.linspace(55,95,100)) 
number_of_mutations=[0,1,2,3,4,5,6]


t_min=50
t_max=75


from scipy import integrate
from scipy.integrate import quad


pop_size=10**5
N=pop_size
top_10_mu = 3.77e-05
u=top_10_mu
alpha=2 #for the double mutant
f_max=0.4999999


#average number of double mutant clone detected in a person
def prob_double_mutant_function(t, VAF_lim, f_max, s): 
    f_min=VAF_lim
    n_star=(np.exp(alpha*s*t)-1)/(alpha*s)
    f_star=n_star/(2*(N+n_star))
    if VAF_lim<f_star:
        if f_max>f_star:
            f_max=f_star
    f_spectrum= lambda f: (2*N/(1-2*f))*(1+(2*f/(1-2*f)))*(alpha*N*u**2/(alpha*s*(2*N*f/(1-2*f))+1))*((np.exp(s*t)/(alpha*s*(2*N*f/(1-2*f))+1)**(1/alpha))-1)
    result = integrate.quad(f_spectrum, f_min,f_max)[0]
    
    if VAF_lim>f_star:
        result=0

    return result

#average number of separate mutations detected in a person
def prob_multiple_singles_function(t, VAF_lim, f_max, s): 
    n_tilde=(np.exp(s*t)-1)/s
    f_min=VAF_lim
    n_star=(np.exp(alpha*s*t)-1)/(alpha*s)
    f_star=n_star/(2*(N+n_star))
    if VAF_lim<f_star:
        if f_max>f_star:
            f_max=f_star
    f_spectrum= lambda f: (2*N/(1-2*f))*(1+(2*f/(1-2*f)))*N*u/(2*N*f/(1-2*f))*np.exp(-(2*N*f/(1-2*f))/n_tilde)
    result = integrate.quad(f_spectrum, f_min,f_max)[0]
    
    if VAF_lim>f_star:
        result=0

    return result

def Prob_separate(k, VAF_lim):
    b = 0.33
    d = 3.29e-04
    s_max = 0.161
    s_peak=0


    weight_s= lambda s: np.exp(-((abs(s-s_peak))/d)**b)
    normalization=integrate.quad(weight_s, 0.0, s_max)


    #before_wrt_l_log= lambda t, l, s: pop_size/(t_max-t_min)*u*np.exp(l)*((1.0/normalization[0])*np.exp(-((abs(s-s_peak))/d)**b))/np.exp(l)*np.exp(-np.exp(l)/((np.exp(s*t)-1)/s))
    before_wrt_l_log= lambda t, s: ((1.0/normalization[0])*(1/(t_max-t_min))*np.exp(-((abs(s-s_peak))/d)**b))\
    *(np.exp(-prob_multiple_singles_function(t, VAF_lim, f_max, s))*(prob_multiple_singles_function(t, VAF_lim, f_max, s))**k/(math.factorial(k)))
    result = integrate.dblquad(before_wrt_l_log, 0, s_max, lambda s: t_min, lambda s: t_max)[0]
    
    return result


with open(datafilename) as data:
    reader = csv.reader(data,delimiter=',')
    #reader = csv.DictReader(data)
    for row in reader:
        try:
            #cannot assign without elliciting IndexError 
            Participant_ID= row[1]
            age=float(row[2])
            number_of_mutations= float(row[3])
            

            if row[0] == target_study: 
                if number_of_mutations==6:
                    counts_6_mut=counts_6_mut+1
                elif number_of_mutations==5:
                    counts_5_mut=counts_5_mut+1
                elif number_of_mutations==4:
                    counts_4_mut=counts_4_mut+1
                elif number_of_mutations==3:
                    counts_3_mut=counts_3_mut+1
                elif number_of_mutations==2:
                    counts_2_mut=counts_2_mut+1
                elif number_of_mutations==1:
                    counts_1_mut=counts_1_mut+1
            #or row[8] == subject_gene or row[12] == subject_gene  or row[16] == subject_gene\
            #or row[20] == subject_gene  or row[24] == subject_gene or row[28] == subject_gene or row[32] == subject_gene\
            #or row[36] == subject_gene or row[40] == subject_gene or row[44] == subject_gene or row[48] == subject_gene\
            #or row[52] == subject_gene or row[56] == subject_gene:
#                 if age<60:
#                     base_number_less60=base_number_less60+1
#                 elif 60<=age<70:
#                     base_number_60=base_number_60+1
#                 elif 70<=age<80:
#                     base_number_70=base_number_70+1
#                 elif 80<=age<90:
#                     base_number_80=base_number_80+1
#                 elif age>=90:
#                     base_number_great90=base_number_great90+1
#                 else:
#                     print('age is ', age)
                    

 

                            
        except: # catch the error
                pass # pass will basically ignore it
         # and execution will continue on to whatever comes
         # after the try/except block
#        cc=cc+1
counts_0_mut= total_pop_num- counts_1_mut- counts_2_mut- counts_3_mut- counts_4_mut- counts_5_mut- counts_6_mut
counts=[counts_0_mut, counts_1_mut,counts_2_mut,counts_3_mut,counts_4_mut,counts_5_mut,counts_6_mut]
Divided_counts=[x/total_pop_num for x in counts]
data=[Divided_counts[i] for i in x_vector]   #truncated data



#this no longer affects the divided counts
#avoid dividing by 0
if counts[0]==0:
    counts[0]=1
if counts[1]==0:
    counts[1]=1
if counts[2]==0:
    counts[2]=1
if counts[3]==0:
    counts[3]=1
if counts[4]==0:
    counts[4]=1
if counts[5]==0:
    counts[5]=1
if counts[6]==0:
    counts[6]=1


Errors=[Divided_counts[0]/(counts[0])**0.5,Divided_counts[1]/(counts[1])**0.5,Divided_counts[2]/(counts[2])**0.5\
        ,Divided_counts[3]/(counts[3])**0.5,Divided_counts[4]/(counts[4])**0.5,Divided_counts[5]/(counts[5])**0.5,\
Divided_counts[6]/(counts[6])**0.5]
#mu= lambda VAF_lim: [Prob_separate(k, VAF_lim) for k in x_vector]    #expected probability
#manually enter error
sigma=[Divided_counts[0]/(counts[0])**0.5,Divided_counts[1]/(counts[1])**0.5,Divided_counts[2]/(counts[2])**0.5\
        ,Divided_counts[3]/(counts[3])**0.5,Divided_counts[4]/(counts[4])**0.5]

def log_likelihood(VAF_lim):
    mu=[Prob_separate(k, VAF_lim) for k in x_vector] 
    before_sum= [-num_data_pts*np.log(sigma[i])-1/(2*(sigma[i])**2)*(data[i]-mu[i])**2   for i in x_vector ]

    return sum(before_sum)

from scipy import optimize
from scipy.optimize import minimize
bounds=[[0,0.0037]]
result = minimize(lambda VAF_lim: -log_likelihood(VAF_lim), 0.1, bounds=bounds, method='TNC')
best_fit_VAF_lim = result.x[0]

number_of_mutations=[0,1,2,3,4,5,6]
Prob_multiple_mutations = [Prob_separate(k, best_fit_VAF_lim) for k in number_of_mutations]



#ax1[1][1].scatter(timepoints,Divided_counts, alpha=1.0, color=qualitative_10[3],                marker='.', s=100)
ax1.bar(number_of_mutations, Prob_multiple_mutations, color='goldenrod', label='best-fit VAF limit at %s' % best_fit_VAF_lim)
ax1.errorbar(number_of_mutations,Divided_counts, yerr=Errors, alpha=1.0, color=qualitative_10[3]  , fmt='o' ,
         mec=qualitative_10[3], mfc=qualitative_10[3], ms=15, capsize=15, mew=2, label='Data')
ax1.legend()
#ax1[0][0].plot([phi60 for i in range(100)], [10**(j/50 - 1) for j in range(100)], lw=2, linestyle = ':', color='k')
ax1.set_yscale('log')
ax1.set_xlabel('Number of mutations')
ax1.set_ylabel('Density')
ax1.set_title('Young 2016')
#ax1[0][0].set_xlim(xmin, xmax)
ax1.set_ylim(1/total_pop_num/100,10**1.5)                                                
ax1.set_xticks(number_of_mutations);                                                                                            
ax1.set_xticklabels(number_of_mutations);

#print(Base_numbers)    #number of ppl having the a mutation in the subject gene
#print(Divided_counts)   #number of ppl having more than just a mutation in the subject gene/Base numbers
print(Errors)
Errors=[]


#plt.savefig('Dist_num_of_mutations_Young_2016_age_best_fit.pdf', bbox_inches='tight')

#num_data_pts=5
#x_vector=range(num_data_pts)
#data=[x/total_pop_num for x in x_vector]
```
